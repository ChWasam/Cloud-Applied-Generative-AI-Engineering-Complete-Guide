```python
from openai import OpenAI
from dotenv import load_dotenv,find_dotenv

_:bool = load_dotenv(find_dotenv())

client: OpenAI = OpenAI()
```

```python
from openai.types.chat.chat_completion import ChatCompletion

def chat_completion (prompt:str) -> ChatCompletion:
    response:ChatCompletion = client.chat.completions.create(
        messages = [
            {
                "role" : "user",
                "content": prompt
            }
        ],
        model = "gpt-3.5-turbo-1106",
        stream = True
    )
    for msg in response:
        print(msg)

output : ChatCompletion = chat_completion("The temprature today is 60 C. Explain to me in 3 lines that what i can do if i have no water and i am directly exposed to sun rays. ")
```

```python
from openai.types.chat.chat_completion import ChatCompletion

def chat_completion (prompt:str):
    stream:ChatCompletion = client.chat.completions.create(
        messages = [
            {
                "role" : "user",
                "content": prompt
            }
        ],
        model = "gpt-3.5-turbo-1106",
        stream = True
    )
    for part in stream:
        print(part.choices[0].delta.content or " ")


chat_completion("The temprature today is 60 C. Explain to me in 3 lines that what i can do if i have no water and i am directly exposed to sun rays. ")
# Output in stream 
```

```python
from openai.types.chat.chat_completion import ChatCompletion

stream = client.chat.completions.create(
    messages = [
        {
            "role" : "user",
            "content": "The temprature today is 60 C. Explain to me in 3 lines that what i can do if i have no water and i am directly exposed to sun rays. "
        }
    ],
    model = "gpt-3.5-turbo-1106",
    stream = True
)
for part in stream:
    print(part.choices[0].delta.content or " ", end= "...")

# Output in stream 
```
